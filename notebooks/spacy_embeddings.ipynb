```python
word_embed_generator = spacy.load("en_core_web_md")
```

```python
hello_vec = word_embed_generator("hello").vector
world_vec = word_embed_generator("world").vector
hello_world_vec = word_embed_generator("hello world").vector

print(hello_world_vec)
```
